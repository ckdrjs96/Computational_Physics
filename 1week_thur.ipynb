```
#1 alpha에 따라서 (alpha,1)를 단위 백터를 사용하여 일직선 그리기
```

```
alpha=2
dx=alpha/np.sqrt(alpha**2+1)
dy=1/np.sqrt(alpha**2+1)
plt.arrow(0,0,dx,dy,head_width=0.2)
plt.xlim(-2,2)
plt.ylim(-2,2)
```

```
#2 A=[[1,alpha],[0,alpha]]
#이 연산자를 u=[1,1]애 적용했을때 생성되는 백터들을 그려라
u= np.array([1,1])
for alpha in np.arange(-1,1,0.2):
  A=np.array([[1,alpha],
              [0,alpha]])
  v=np.dot(A,u)
  plt.arrow(0,0,v[0],v[1],head_width=0.1, color='r')

plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```

```
# 1)
# x_A = a1*t^2/2
# x_B= 100 - a2*t^2/2

# x_A = x_B -> a1*t^2/2=100 - a2*t^2/2 ,t=30
#              a2*t^2/2=100 + a2*t^2/2 ,t=90
```

```
t1=30
t2=90
A=np.array([[t1**2,t1**2],
            [t2**2,-1*(t2**2)]])
b=np.array([200,200])
np.linalg.solve(A,b)
```
