```python
#1
d=[2,4,4,2,3]
A=np.diag(d)-np.eye(5,k=1)-np.eye(5,k=-1)
A[2,4]=-2
A[4,2]=-2
k=2.5
w=1
b=np.array([w/k]*5)
print(A,b)
```

```python
def conjGrad(n,A,x,b,tol=1.0e-9):
  r=b-A@x
  s=r.copy()
  
  for i in range(n):
    alpha=(s.T@r)/(s.T@A@s)
    x=x+alpha*s
    r=b-A@x
    # print(f'iter:{i}, x:{x}')
    if (math.sqrt(r@r)) < tol:
      # print(math.sqrt(r@r))
      break
    else:
      beta=-1*(r.T@A@s)/(s.T@A@s)
      s=r+beta*s
  return x,i
```

```python
x=np.zeros(5)
conjGrad(100,A,x,b)
```

```python
n=10
A=4*np.eye(n)-np.eye(n,k=1)-np.eye(n,k=-1)
b=5*np.ones(n)
b[0]=9
```

```python
#conjugate Gradient
x=np.zeros(n)
start=time.time()
ans=conjGrad(100,A,x,b)
print(time.time()-start)
print(ans)
```

```python
#gaussElimin
def gaussElimin(a, b):
    n = len(b)
    # Elimination Phase
    for k in range(0, n-1):
        for i in range(k+1, n):
            if a[i, k] != 0.0:
                lam = a[i,k]/a[k,k]
                a[i, k+1:n] = a[i, k+1:n] - lam * a[k, k+1:n]
                b[i] = b[i] - lam * b[k]

    # Back substitution
    for k in range(n-1, -1, -1):
        b[k] = (b[k] - np.dot(a[k,k+1:n], b[k+1:n]))/a[k,k]
    return b
```

```python
start=time.time()
ans=gaussElimin(A,b)
print(time.time()-start)
print(ans)
```

gaussElimin 방식과 conjugate decent 방식 과 답이 소수셋째자리 까지 일치 하였고 시간은 n=10일때 비슷하였다.

```python
n=10000
A=4*np.eye(n)-np.eye(n,k=1)-np.eye(n,k=-1)
b=5*np.ones(n)
b[0]=9

#conjugate Gradient
x=np.zeros(n)
start=time.time()
ans=conjGrad(100,A,x,b)
print('conjugate:',time.time()-start)
print(ans)

start=time.time()
ans=gaussElimin(A,b)
print('gaussElimin:',time.time()-start)
print(ans)
```
