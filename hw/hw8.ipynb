```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt
```

```python
#1
u=2510
m0=2.8*10**6
m=13.3*10**3
g=9.81
v=335*2
f=lambda t: u*np.log(m0/(m0-m*t))-g*t-v
```

```python
x=np.arange(50,150,1)
```

```python
plt.plot(x,f(x))
```

```python
ans=root_scalar(f,bracket=[50,150])
print('t:',ans.root)
```

```python
#2
f2=lambda x: (np.log(x)-(1-1/x))/(np.log(x)+(1-(1/x))/(r-1))-n
r=5/3
n=0.3        
```

```python
x=np.arange(2,20,1)
plt.plot(x,f2(x))
```

```python
ans=root_scalar(f2,bracket=[2,20])
print('T2/T1:',ans.root)
```

```python
#3
def f3(R,setha,C,e,a):
  return C/(1+e*np.sin(setha+a))-R
```

```python
def concat(input):
  C,e,a=input
  to_de=180/np.pi
  c1=f3(6870,-30*to_de,C,e,a)
  c2=f3(6728,0*to_de,C,e,a)
  c3=f3(6615,30*to_de,C,e,a)
  return [c1,c2,c3]
```

```python
from scipy import optimize
ans=optimize.root(concat,[100,1,0.5])
```

```python
print(f'C={ans.x[0]:.4f}, e={ans.x[1]:.4f}, alpha={ans.x[2]:.4f}')
```
