```
#3
ex= np.array([0,1])
for a in np.arange(-1,1.2,0.2):
  A=np.array([[1,a],
              [0,1]])
  v=np.dot(A,ex)
  plt.arrow(0,0,v[0],v[1],head_width=0.1, color='r')

plt.axhline(1,-2,2)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```

```
#5
#rabbit+chicken=40
#4rabbit+2chicken=92
A=np.array([[1,1],
            [4,2]])
b=np.array([40,92])
print('rabbit:{0},chicken:{1}'.format(*np.linalg.solve(A,b)))
```
