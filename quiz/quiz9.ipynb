```python
import numpy as np
from scipy.integrate import quad
```

```python
#1-1
f=lambda x:np.exp(-x*x)
```

```python
a=1
ans=np.sqrt(np.pi)
tol=1e-10
while True:
  new,_=quad(f,-a,a)
  print(f'a:{a}, integrate={new:.10f}')
  if abs(new-ans) < tol:
    break
  a+=1
print(f'np.sqrt(np.pi)={ans}')
```

```python
#1-2
def trapezoid(f,a,b,Iold,k):
    if k == 1:Inew = (f(a) + f(b))*(b - a)/2.0
    else:
        n = 2**(k -2 ) # Number of new points
        h = (b - a)/n # Spacing of new points
        x = a + h/2.0
        sum = 0.0
        for i in range(n):
            sum = sum + f(x)
            x=x+h
        Inew = (Iold + h*sum)/2.0
    return Inew
```

```python
Iold=0
k=1
tol=1e-10
ans=np.sqrt(np.pi)
while True:
  Iold=trapezoid(f,-10,10,Iold,k)
  print(f'n_panel={2**(k-1)}, integrate={Iold :.10f}')
  if abs(Iold-ans) <=tol:
    break
  k+=1
print(f'np.sqrt(np.pi)={ans}')
```

```python
#3
def w(H,m):
  G=6.673e-11
  M=5.972e24
  R=6371
  f=lambda r: G*M*m/(r*r)
  inte=quad(f,R,R+H)
  return inte[0]
```
