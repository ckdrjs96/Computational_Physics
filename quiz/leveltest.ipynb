```
#2
a=int(input())
b=int(input())
def LCM(a,b):
  prime=[]
  while True:
    for i in range(2,b+1):
      if a%i==0 and b%i==0:
        prime.append(i)
        a=a//i
        b=b//i
        break
    else:
      prime+=[a,b]
      return prime
    if a==1 or b==1:
      prime+=[a,b]
      return prime
ans=1
for i in LCM(a,b):
  ans*=i
print(ans)
```

```
#3
a=int(input())
b=int(input())
def LCM(a,b):
  prime=[]
  while True:
    for i in range(2,b+1):
      if a%i==0 and b%i==0:
        prime.append(i)
        a=a//i
        b=b//i
        break
    else:
      prime+=[1]
      return prime

    if a==1 or b==1:
      return prime

ans=1
for i in LCM(a,b):
  ans*=i
print(ans)
```
