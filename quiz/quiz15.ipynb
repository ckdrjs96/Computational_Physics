```python
import numpy as np
from scipy.integrate import solve_ivp, solve_bvp
import matplotlib.pyplot as plt
```

```python
#1
def fun(x,y):
  f1=y[1]
  f2= 1-y[1]
  return np.array([f1,f2]) #(n,m)

def bc(ya,yb): #(왼쪽, 오른쪽)
  return np.array([ya[0],yb[0]])

x = np.linspace(0, np.pi/2, 101)
y = np.zeros((2, x.size))
sol=solve_bvp(fun, bc, x, y)
plt.plot(x,sol.sol(x)[0],label='y')
plt.plot(x,sol.sol(x)[1],label='y`')
plt.legend()
```

```python
#2
def fun(x,y):
  f1=y[1]
  f2= (1+4*x**4+4*x**3*y[0]+y[1])/x
  return np.array([f1,f2]) #(n,m)

def bc(ya,yb): #(왼쪽, 오른쪽)
  return np.array([ya[0],yb[0]-1])

x = np.linspace(1, 2, 40)
y = np.zeros((2, x.size))
sol=solve_bvp(fun, bc, x, y)
plt.plot(x,sol.sol(x)[0],label='y')
plt.plot(x,sol.sol(x)[1],label='y`')
plt.legend()
```

$\delta T/\delta t=\delta t(t)/\delta t *X(x)=t'(t)X(x)$  
$c^2\delta T/\delta x^2=c^2*t(t)* \delta X(x)/\delta x^2=c^2X''(x)t(t)$  
$ c^2X''(x)/X(x)=t'(t)/t(t)=-\lambda^2$


```python
#4-2
landa=2
def fun(x,y):
  f1=y[1]
  f2= -landa**2*y[0]
  return np.array([f1,f2]) #(n,m)

def bc(ya,yb): #(왼쪽, 오른쪽)
  return np.array([ya[1]-100*landa,ya[0]])

for landa in np.arange(-5,0,0.01):
  x = np.linspace(0, 2, 40)
  y = np.zeros((2, x.size))
  sol=solve_bvp(fun, bc, x, y)
  if abs(sol.sol(2)[0])<1e-2:
    print('ha',landa)
# plt.plot(x,sol.sol(x)[0],label='y')
# plt.plot(x,sol.sol(x)[1],label='y`')
# plt.legend()
```
