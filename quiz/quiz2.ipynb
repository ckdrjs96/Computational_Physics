```python
def LU(a,b):
    n=len(b)
    A=a.copy()
    L=np.eye(n)
    for j in range(0,n-1):
      for i in range(j+1,n):
        if A[i,j] !=0:
          L[i,j]=A[i,j]/A[j,j]
          A[i]=A[i]-L[i,j]*A[j]
    
    U=A.copy()
    return L,U
```

```python
L,U=LU(A,b)
```
