```python
#2
A=np.array([[3.,2],
            [2,6]])
b=np.array([2,8])
x0=np.array([1,2])
```

```python
r0=b-A@x0
s0=r0
a0=(s0.T@r0)/(s0.T@A@s0)
x1=x0+a0*s0

r1=b-A@x1
b0=-1*(r1.T@A@s0)/(s0.T@A@s0)
s1=r1+b0*s0
a1=(s1.T@r1)/(s1.T@A@s1)
x2=x1+a1*s1

```

```python
print(f'r0:{r0},s0:{s0}, alpha0:{a0:.4}, beta0:{b0:.4}\nr1:{r1},s1:{s1},x1:{x1},x2:{x2}')
```

```python
#3
def ConjGrad(A,x,b,tol=1.0e-9):
    n = len(b)
    r = b - A@x
    s = r.copy()
    for i in range(n):
        u = A@s
        alpha = np.dot(s,r)/np.dot(s,u)
        x = x + alpha*s
        r = b - A@x
        if(np.sqrt(np.dot(r,r))) < tol:
            break 
        else:
            beta = -np.dot(r,u)/np.dot(s,u)
            s = r + beta*s
    
    return x
```
