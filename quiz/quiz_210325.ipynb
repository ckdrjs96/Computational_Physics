```python
#2-3
def forwardsub(A,b):
  n=len(b)
  for k in range(0,n):
    b[k]=(b[k]-np.dot(A[k,:k],b[:k]))/A[k,k]
  return b
```

```python
#3-1
A=np.array([[0.95,0.02],
          [0.05,0.98]])
```

```python
#3-2
lam1=A[1,0]/A[0,0]
A[1]=A[1]-lam1*A[0]
```

```python
U=A.copy()
```

```python
L=np.eye(2)
L[1,0]=lam1
```

```python
#3-3
def ans(L,U,b):
#L@y=b
  y=np.zeros(2)
  y[0]=b[0]
  y[1]=b[1]-b[0]*L[1,0]

#U@x=y
  x=np.zeros(2)
  x[1]=y[1]/U[1,1]
  x[0]=(y[0]-y[1]*U[0,1])/U[0,0]
  return x
```

```python
b=np.array([30,24])
```

```python
b2020=ans(L,U,b)
b2019=ans(L,U,b2020)
b2018=ans(L,U,b2019)
```
