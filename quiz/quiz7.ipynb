```python
#2-1
f=lambda x: x**3-2*x**2-2*x+1
df=lambda x: 3*x**2-4*x-2
df2=lambda x: 6*x-4
```

```python
x=np.arange(-1.5,3,0.1)
plt.plot(x,df(x))
```

```python
from bisection import *
start=time.time()
bi_ans=bisection(df,1,2,switch=1,tol=1e-5)
end=time.time()
print('time:',end-start, 'x`=',bi_ans,'f(x`)=',f(bi_ans))
```

```python
def newton(f,df,x):
  return x-f(x)/df(x)
```

```python
x=1
start=time.time()
for i in range(100):
  x=newton(df,df2,x)
  if abs(bi_ans-x)<1e-5:
    net_ans=x
    break
end=time.time()
print('time:',end-start, 'x`=',net_ans,'f(x`)=',f(net_ans))
```

```python
#3
lim=4.0
delta=0.1
xdata=np.arange(-lim,lim,delta)
ydata=xdata.copy()
x,y=np.meshgrid(xdata,ydata)
def f1(x,y):
  return (x-2)**2+y**2-4

def f2(x,y):
  return x**2+(y-2)**2-4

F1=f1(x,y)
F2=f2(x,y)
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_aspect('equal')
plt.contour(x,y,F1,[0])
plt.contour(x,y,F2,[0])
```

```python
from scipy.optimize import root
def func(x):
  return [f1(*x),f2(*x)]

x0=[1,0]
sol=root(func,x0)
print(f'(x1,y1)=({sol.x[0]:.3f},{sol.x[1]:.3f})')
```

```python
x0=[3,3]
sol=root(func,x0)
print(f'(x2,y2)=({sol.x[0]:.3f},{sol.x[1]:.3f})')
```
