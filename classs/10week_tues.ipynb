```python
from scipy.interpolate import CubicSpline
```

```python
f=lambda x:np.exp(x)+2*np.exp(x**2)
derivative(f,1,n=1,dx=1e-6)
```

```python
#나}
x=np.arange(0,2,0.2)
y=f(x)
```

```python
cs=CubicSpline(x,y)
```

```python
plt.plot(x,cs(x))
```

```python
cs1=cs.derivative(1)
```
