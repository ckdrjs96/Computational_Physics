```python
def f(t,psi):
  psi0,psi1 =psi
  f1=psi1
  f2=-w**2*psi0
  return [f1,f2]
```

```python
#가
w=1
sol1=solve_ivp(f,[0,10],[0,w])
w=2
sol2=solve_ivp(f,[0,10],[0,w])
w=3
sol3=solve_ivp(f,[0,10],[0,w])
```

```python
plt.plot(sol1.t,sol1.y[0],label='w=1')
plt.plot(sol2.t,sol2.y[0],label='w=2')
plt.plot(sol3.t,sol3.y[0],label='w=3')
plt.legend()
```

```python
def f(t,psi,w):
  psi0,psi1 =psi
  f1=psi1
  f2=-w**2*psi0
  return [f1,f2]
t_span=[0,10]
yini=[0,w]
t_eval=np.linspace(0,10,30)
w=1
sol=solve_ivp(f,t_span,yini,t_eval=t_eval,args=[w])
```

```python
plt.plot(sol.t,sol.y[0])
```

```python
def error(w):
  sol=solve_ivp(f,t_span,[0,w],t_eval=t_eval,args=[w])
  error=sol.y[0][-1]
  return error
```

```python
error(1)
```

```python
warray=np.linspace(0,1,100)
vfun=np.vectorize(error)
plt.plot(warray,vfun(warray))
```
