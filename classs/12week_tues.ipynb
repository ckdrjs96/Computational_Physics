```python
def simp(f,a,b,tol):
  Iold=0
  for n in range(2,10000):
    x=np.linspace(a,b,2*n+1)
    y=f(x)
    h=x[1]-x[0]
    I1=h*(y[0]+y[-1])
    I4=4*h*sum(y[1::2])
    I2=2*h*sum(y[2:-2:2])
    Itot=(I1+I4+I2)/3
    if (k>1) and (abs(Itot-Iold)<tol):break
    Iold=Itot
  return Itot,2*n+1

  print(x,y)
```
