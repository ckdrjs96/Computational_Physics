```python
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import numpy as np
```

```python
f=lambda x :(x-2)*x*(x+2)**2
```

```python
x=np.arange(-3,2,0.001)
plt.plot(x,f(x))
```

```python
minimize(f,0)
```

```python
minimize(f,[1],bounds=[(-1.5,2)])
```

```python
def ff(x):
  return x[0]**2.0+x[1]**2

x0=[1,1]
minimize(ff,x0)
```

```python
def area(x):
  a=x[0]
  b=x[1]
  return a*b*4 +a*a
def vol(x):
  return x[0]*x[0]*x[1]-1

cons =[{'type':'eq','fun':vol}]
x0=[1,1]
minimize(area,x0,constraints=cons)
```
