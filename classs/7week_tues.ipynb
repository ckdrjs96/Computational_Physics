```python
def coeffts(xData,yData):
    m = len(xData)  # Number of data points
    a = yData.copy()
    for k in range(1,m):
        a[k:m] = (a[k:m] - a[k-1])/(xData[k:m] - xData[k-1])
    return a
```

```python
def neville(xData, yData, x):
    m = len(xData) # number of data points
    y = yData.copy()
    for k in range(1, m):
        y[0:m-k] = ((x-xData[k:m])*y[0:m-k] + 
                (xData[0:m-k] - x)*y[1:m-k+1])/(xData[0:m-k] - xData[k:m])

    return y[0]
```
