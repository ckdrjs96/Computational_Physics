```
import numpy as np
import time
A_list=np.ones((3000,3000)).tolist()
A_np=np.ones((3000,3000))

start_time=time.time()
for i in range(3000):
  for j in range(3000):
    A_np[i,j]=A_np[i,j]+1
end_time=time.time()
print(end_time-start_time)
```
