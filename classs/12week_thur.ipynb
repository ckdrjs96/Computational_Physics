```python
'''
y``=f(x,y,y`)
y0=y, y1=y`

y0`=y1
y1`=f(x,y0,y1)

y``=x-2y
x``=y-x
'''
def f(t,y):
  y0,y1,x0,x1=y
  # f=F(t,y,y`,x,x`), y`=f
  f1=y1
  f2=x0-2*y0
  f3=x1
  f4=y0-x0
  return [f1,f2,f3,f4]
```

```python
g=9.80665
cd=0.2028
m=80

def f(t,y):
  y0,y1 =y
  f1 = y1
  f2 = g-(cd/m)*y1**2

  return [f1,f2]
```

```python
y0=[0,0] #[y0,y1]
tsapn=[0,50]
sol=solve_ivp(f,tsapn,y0)
```

```python
import matplotlib.pyplot as plt
plt.plot(sol.t,sol.y[0])
```
