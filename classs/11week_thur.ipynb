```python
import numpy as np
```

```python
#사다리꼴 적분
x=np.array([0,0.5,1.0,1.5,2])
y=np.array([1.5,2,2,1.6364,1.25])
```

```python
#trapezoid 1/2 rule weights=1 2 2 2 1
h=.5
h*(y[0]+2*y[1]+2*y[2]+2*y[3]+y[4])*(1/2)
```

```python
#cubic spline
from scipy.interpolate import CubicSpline
from scipy.integrate import *
cs= CubicSpline(x,y)
quad(cs,0,2)
```

```python
#사다리꼴 적분 3/8rule
x=np.array([0,0.5,1.0,1.5,2,2.5])
y=np.array([1.5,2,2,1.6364,1.25,0.9565])
```

```python
#사다리꼴 1/2
I=h*(2*sum(y)-(y[0]+y[-1]))/2
print(I)
```

```python
#simpson 1/3rule 1 4 2 4 1
I1=h*(y[0]+y[-1])
I4=4*h*sum(y[1::2])
I2=2*h*sum(y[2:-2:2])
Itot=(I1+I4+I2)/3
```

```python
#첫째 패널 사다리꼴 나머지 simpson
I1=(y[0]+y[1])*(h/2)
I2=(y[1]+4*y[2]+2*y[3]+4*y[4]+y[5])*(h/3)
print(I1+I2)
```

```python
# 처음 2개 simpson , 나머지 simpson 3/8
I1=(y[0]+4*y[1]+y[2])*(h/3)
I2=(y[2]+3*y[3]+3*y[4]+y[5])*(3*h/8)
print(I1+I2)
```
