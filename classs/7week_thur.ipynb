```python
from scipy.optimize import root_scalar
import numpy as np

def f(x):
  return x/2-np.sin(x)

root_scalar(f,method='brentq',bracket=[-1,0])
```

```python
def f2(x):
  return x*np.sin(x)+3*np.cos(x)-x

root_scalar(f2,method='brentq',bracket=[-6,-4])
```

```python
root_scalar(f2,method='brentq',bracket=[-4,0])
```

```python
root_scalar(f2,method='brentq',bracket=[0,6])
```

```python
root_scalar(f2,method='ridder',bracket=[-6,6])
```

```python
def h(lamda):
  L=160
  return (np.cosh(lamda*L/2)-1)/lamda-15

root_scalar(h,method='ridder',bracket=[-6,5])
```

```python
h_ans=root_scalar(h,method='brentq',bracket=[-6,5])
```

```python
lamda=h_ans.root
L=160
s=(2/lamda)*np.sinh(lamda*L/2)
s
```

```python
import matplotlib.pyplot as plt

x=np.arange(-0.1,0.1,0.0001)
y=h(x)
plt.plot(x,y)
```
