```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#curve fit

x=np.arange(5)+1
y=np.array([3,6,3,6,4])
```

```python
pfit=np.polyfit(x,y,4)
```

```python
pfit
```

```python
#y=0.2*x+3.8
line=np.poly1d(pfit)
```

```python
xx=np.linspace(0,6,50)
yy=line(xx)

plt.plot(x,y,'*') #original
plt.plot(xx,yy) #linear
```

```python
#curve fit(f,xdata,ydata)
from scipy.optimize import curve_fit

def f(x,m,b):
  return m*x+b

parm,cov_mat=curve_fit(f,x,y)

yy1=f(xx,parm[0],parm[1])

plt.plot(x,y,'*') #original
plt.plot(xx,yy) #linear
plt.plot(xx,yy1,'--') #second linear fit
```

```python
x=np.array([-2,1,4,-1,3,-4])
y=np.array([-1,2,59,4,24,-53])
```

```python
def f2(x,b2,b1,b):
  return b2*x**2+b1*x+b

parm,cov_mat=curve_fit(f2,x,y)


xx=np.linspace(-4,6,100)
yy=f2(xx,*parm)
plt.plot(x,y,'*') #original
plt.plot(xx,yy) #quadra
```

```python
data = np.loadtxt('oildata.dat',delimiter=',')
```

```python
x,y=data[0],data[1]

def f3(x,a,b,c,d):
  return a*np.log(b*x+c)+d

def f4(t,v0,tau):
  return v0*(1-np.exp(-t/tau))

parm,cov_mat=curve_fit(f3,x,y)
 
xx=np.linspace(0,1.3,50)
yy=f3(xx,*parm)
plt.plot(x,y,'*') #original
plt.plot(xx,yy,'g',label='log') #log

parm,cov_mat=curve_fit(f4,x,y)
 
xx=np.linspace(0,1.3,50)
yy=f4(xx,*parm)
# plt.plot(x,y,'*') #original
plt.plot(xx,yy,'r',label='exp') #exp
plt.legend()
plt.show()
```
