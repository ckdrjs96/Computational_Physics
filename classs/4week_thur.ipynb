```python
import numpy as np

c=np.array([-2,-6,12,-6],dtype=float)
d=np.array([2,5,16,39,14],dtype=float)
e=c.copy()

b=np.array([1,0,0,0,0],dtype=float)
amat=np.diag(c,k=-1)+np.diag(d,k=0)+np.diag(e,k=1)
```

```python
np.linalg.solve(amat,b)
```
