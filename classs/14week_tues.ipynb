```python
import numpy as np
from scipy.integrate import solve_ivp, solve_bvp
import matplotlib.pyplot as plt
```

y''+k**2+y=0  
y(0)=y(1)=0 (경계조건)

```python
'''
y'(0)=k
y''=-k**2*y
'''
def fun(x,y,p):
  k=p[0]
  f1=y[1]
  f2= -k**2 *y[0]

  return np.array([f1,f2]) #(n,m)

def bc(ya,yb,p): #(왼쪽, 오른쪽)
  k=p[0]
  return np.array([ya[0],yb[0],ya[1]-k])


```

```python
x = np.linspace(0, 1, 5)
y = np.zeros((2, x.size))
y[0, 1] = 1
y[0, 3] = -1
```

```python
sol = solve_bvp(fun, bc, x, y, p=[6])
```
