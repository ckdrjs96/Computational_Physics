```python
#3X3
# c=np.ones(2)
# e=2*np.ones(2)
# d=4*np.ones(3)
n=10
c=np.ones(n-1)
e=2*np.ones(n-1)
d=4*np.ones(n)

D=np.diag(d)
E=np.diag(e,k=1)
C=np.diag(c,k=-1)
A=C+D+E
b=np.ones(n)
```

```python
D=np.diag(d)
D
```

```python
E=np.diag(e,k=1)
E
```

```python
C=np.diag(c,k=-1)
C
```

```python
A=C+D+E
A
```

```python
y=np.zeros(n)
x=np.zeros(n)

for k in range(1,n): #두번째 행부터 끝까지
  lam=c[k-1]/d[k-1]
  d[k] = d[k]-lam*e[k-1]
  c[k-1]=lam

#L@y=b forward
y[0] = b[0]
for k in range(1,n):
  y[k]=b[k]-c[k-1]*y[k-1]

#U@x=y backward
x[n-1]=y[n-1]/d[n-1]
for k in range(n-2,-1,-1):
  x[k]=(y[k]-e[k]*x[k+1])/d[k]
```

```python
x
```

```python
np.linalg.solve(A,b)
```
